```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('double-firm_varying_framework_duration.csv', error_bad_lines=False, header=6)  
df = df.replace(['<RuntimePrimitiveException>'], 0.0)
df['mean BANKRUPT-FIRM-PROFITS'] = df['mean BANKRUPT-FIRM-PROFITS'].astype(float)
df['firm-turnover'] = df['TOTAL-BANKRUPT-FIRMS / n-firms'] / 200


display(df)
```

```python
ax = sns.scatterplot(x= "framework-duration", 
                     y= "firm-turnover", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. firm-turnover")
plt.show()

ax = sns.scatterplot(x= "framework-duration", 
                     y= "mean UNEMPLOYMENT-RATES", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. unemployment")
plt.show()

ax = sns.scatterplot(x= "framework-duration", 
                     y= "mean MEAN-PRICES", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. price")
plt.show()

ax = sns.scatterplot(x= "framework-duration", 
                     y= "standard-deviation MEAN-PRICES", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. price volatility")
plt.show()

ax = sns.scatterplot(x= "framework-duration", 
                     y= "mean BANKRUPT-FIRM-PROFITS", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. Bankrupt Firm Profits")
plt.show()

ax = sns.scatterplot(x= "framework-duration", 
                     y= "mean [lifetime-profits] of firms", 
                     data = df)
plt.title("Two-Layer Firm: framework-duration vs. Active Firm Profits")
plt.show()
```
