```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('single-firm_varying_memory_constant.csv', error_bad_lines=False, header=6)  
df = df.replace(['<RuntimePrimitiveException>'], 0.0)
df['mean BANKRUPT-FIRM-PROFITS'] = df['mean BANKRUPT-FIRM-PROFITS'].astype(float)
df['firm-turnover'] = df['TOTAL-BANKRUPT-FIRMS / n-firms'] / 200


display(df)
```

```python
ax = sns.scatterplot(x= "firm-memory-constant", 
                     y= "firm-turnover", 
                     data = df)
plt.title("One-Layer Firm: firm-memory-constant vs. firm-turnover")
plt.show()

ax = sns.scatterplot(x= "firm-memory-constant", 
                     y= "mean UNEMPLOYMENT-RATES", 
                     data = df)
plt.title("One-Layer Firm: firm-memory-constant vs. unemployment")
plt.show()

ax = sns.scatterplot(x= "firm-memory-constant", 
                     y= "mean MEAN-PRICES", 
                     data = df)
plt.title("One-Layer Firm: firm-memory-constant vs. price")
plt.show()

ax = sns.scatterplot(x= "firm-memory-constant", 
                     y= "mean ALL-FIRM-PROFITS", 
                     data = df)
plt.title("One-Layer Firm: firm-memory-constant vs. ALL FIRM PROFITS")
plt.show()

ax = sns.scatterplot(x= "firm-memory-constant", 
                     y= "mean BANKRUPT-FIRM-PROFITS", 
                     data = df)
plt.title("One-Layer Firm: firm-memory-constant vs. BANKRUPT FIRM PROFITS")
plt.show()
```
