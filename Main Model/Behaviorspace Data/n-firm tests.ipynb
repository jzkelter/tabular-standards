```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('multifirm_baseline baseline_varying_firm_number.csv', error_bad_lines=False)
df['n-firms'] = df['n-firms'] 
df['firm-turnover'] = df['[mean]'] / df['n-firms']
df['max-firm-turnover'] = df['[max]'] / df['n-firms']

display(df)
ax = sns.scatterplot(x= "n-firms", 
                     y= "firm-turnover", 
                     data = df)
plt.title("Two-Layer Firm: n-firms vs. mean firm turnover")
plt.show()
```

```python
ax = sns.scatterplot(x= "n-firms", 
                     y= "max-firm-turnover", 
                     data = df)
plt.title("Two-Layer Firm: n-firms vs. max firm turnover")
plt.show()
```
