```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('single-firm_varying_firm_number.csv', error_bad_lines=False, header=6)  
df['firm-turnover'] = df['TOTAL-BANKRUPT-FIRMS / n-firms'] / 400
df['unemployment'] = 1 - df['count households with [employed?]'] / 1000


display(df)
```

```python
ax = sns.scatterplot(x= "n-firms", 
                     y= "firm-turnover", 
                     data = df)
plt.title("One-Layer Firm: n-firms vs. mean firm turnover (n-households = 500)")
plt.show()

ax = sns.scatterplot(x= "n-firms", 
                     y= "unemployment", 
                     data = df)
plt.title("One-Layer Firm: n-firms vs. unemployment (n-households = 500)")
plt.show()

ax = sns.scatterplot(x= "n-firms", 
                     y= "mean [price] of firms", 
                     data = df)
plt.title("One-Layer Firm: n-firms vs. price (n-households = 500)")
plt.show()

ax = sns.scatterplot(x= "n-firms", 
                     y= "mean ALL-FIRM-PROFITS", 
                     data = df)
plt.title("One-Layer Firm: n-firms vs. ALL FIRM PROFITS (n-households = 500)")
plt.show()

ax = sns.scatterplot(x= "n-firms", 
                     y= "mean BANKRUPT-FIRM-PROFITS", 
                     data = df)
plt.title("One-Layer Firm: n-firms vs. BANKRUPT FIRM PROFITS (n-households = 500)")
plt.show()
```
